## 1. Union-Find Basics

### Operations
- **Find(x)**: Find root of x's set
- **Union(x, y)**: Merge sets containing x and y

### Optimizations
- **Path compression**: Point directly to root
- **Union by rank**: Attach smaller tree under larger

With both: O(Œ±(n)) per operation (nearly constant)
