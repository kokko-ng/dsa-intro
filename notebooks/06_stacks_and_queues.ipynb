## 1. Stack & Queue Basics

### Stack (LIFO)
```python
stack = []
stack.append(x)  # Push - O(1)
stack.pop()      # Pop - O(1)
stack[-1]        # Peek - O(1)
```

### Queue (FIFO)
```python
from collections import deque
queue = deque()
queue.append(x)   # Enqueue - O(1)
queue.popleft()   # Dequeue - O(1)
```
