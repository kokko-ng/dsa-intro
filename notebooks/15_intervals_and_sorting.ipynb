## 1. Interval Patterns

### Common Approach
1. Sort intervals by start (or end) time
2. Process sequentially, tracking current state

### Key Conditions
- **Overlap**: a.end > b.start (when sorted by start)
- **Merge**: Combine overlapping intervals
